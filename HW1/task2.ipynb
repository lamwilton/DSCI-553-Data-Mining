```python
from pyspark import SparkContext
import json
import sys


review = []
business = []
with open('review.json', encoding='utf-8') as review_file:
    for jsonobj in review_file:
        review_line = json.loads(jsonobj)
        filter = dict({'review_id': review_line['review_id'], 'business_id': review_line['business_id'], 'stars': review_line['stars']})
        review.append(filter)
review
```

```python
with open('business.json', encoding='utf-8') as bus_file:
    for jsonobj in bus_file:
        bus_line = json.loads(jsonobj)
        filter = dict({'business_id': bus_line['business_id'], 'categories': bus_line['categories']})
        business.append(filter)
business
```
